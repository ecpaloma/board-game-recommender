```python
# Import libraries
import pandas as pd
import numpy as np
from ast import literal_eval
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
# read the data
df = pd.read_csv("bgg_dataset.csv", delimiter=";", decimal=",")
#get rid of NAs
df.fillna('NA', inplace=True)

### One hot encode Domains
# Join every string in every row, split the result, pull out the unique values.
domains = np.unique(', '.join(df['Domains']).split(', '))
# Drop 'NA'
domains = np.delete(domains, np.where(domains == 'NA'))
for domain in domains:
    df[domain] = df['Domains'].str.contains(domain).astype('int')

df.drop('Domains', axis=1, inplace=True)


df.sample(n=10)
df[df['Mechanics'].str.contains("Cooperative")]
```
